## Standard Scaler:

- mean = 0 and std = 1


```python
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
std_data = s.fit_transform(df) # converts df data into standard scaler distribution data like mean = 0 and std = 1.
std_data
```

```python
std1 = pd.DataFrame(std_data, columns = df.columns)
std1.head()
```

```python
std1.describe()
```

```python
import seaborn as sns

plt.title("After applying Standard scaling ")
sns.kdeplot(std1['TV'])
sns.kdeplot(std1['radio'])
sns.kdeplot(std1['newspaper'])
sns.kdeplot(std1['sales'])
plt.legend()
```
