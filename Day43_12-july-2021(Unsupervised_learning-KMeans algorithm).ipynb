## CLUSTERING ALGORITHM :

- Group of similar datapoints to clulsters.
- kmeans()

***kmeans Clustering***
- k refers to no. of clusters
- To select the optimal value of k, use elbow method.
***Steps involved:**
    - Choose the value of k.
    - Choose the centroids.
    - Calculate the distance and mean.
    - Repeat the steps until you reach the point of convergence

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
iris.keys()
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris_df.head()
```

```python
x = iris.data
```

```python
from sklearn.cluster import KMeans
k = range(1, 11)
l = []
for i in k :
    model = KMeans(n_clusters=i, random_state=42)
    model.fit(x)
    sse = model.inertia_
    l.append(sse)
    print('The value of sse for k={} is {}'.format(i, sse))
```

```python
plt.plot(k, l)
plt.xlabel('no. of clusters')
plt.ylabel('sse')
plt.show()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3, random_state=42)
model.fit(x)
```

```python
center = model.cluster_centers_
center
```

```python
iris_df['target'] = iris.target
```

```python
iris_df
```

```python
iris_df['target'].value_counts()
```

```python
iris_df['pred_target'] = model.predict(x)
y = model.predict(x)
```

```python
iris_df.head()
```

```python
iris_df['pred_target'].value_counts()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x, y)
```
